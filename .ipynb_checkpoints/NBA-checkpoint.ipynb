```python
import numpy as np
import pandas as pd
```

```python
#Salary data for 21-22

data_salary_21_22 = pd.read_csv('NBA Salary 21-22.csv')
print(data_salary_21_22.shape)
data_salary_21_22.info()
data_salary_21_22.head()
```

```python
#Stats data for 21-22

data_stats_21_22 = pd.read_csv('NBA Stats 21-22.csv')
print(data_stats_21_22.shape)
data_stats_21_22.info()
data_stats_21_22.head()
```

```python
# cleaning the data, drop the unnecessary columns and get rid of the '$', and position in name on salary data.
# Then mergy stat data and salary data with players name.

def data_cleaning (salary, stats):
    stats = stats.drop(['RANK'], axis=1)
    stats.rename(columns={'FULL NAME': 'NAME'}, inplace=True)
    salary['NAME'] = salary['NAME'].str.split(',').str[0]
    salary['SALARY'] = salary['SALARY'].replace('\$','',regex=True)
    salary['SALARY'] = salary['SALARY'].replace(',','',regex=True)
    #data = stats.join(salary.set_index('NAME'), on='NAME')
    data = stats.merge(salary, on='NAME', how='left')
    data = data.dropna()
    return data
```

```python
##Data cleaning and concatenation of salary and stat data(21-22)



data_21_22 = data_cleaning(data_salary_21_22, data_stats_21_22)
# add the year columns
data_21_22['YEAR'] = '21-22'
```

## Random Forest with CV

```python
#Random Forest with cross validation

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV

## using GridSearchCV to find best max_features:

import time
num_feature = len(X_train1.columns)
grid_values = {'max_features': np.linspace(1,num_feature,num_feature, dtype='int32'),
               'min_samples_leaf': [5],
               'n_estimators': [500],
               'random_state': [88]} 

tic = time.time()

rf2 = RandomForestRegressor() 
cv = KFold(n_splits=5,random_state=333,shuffle=True) 
rf_cv = GridSearchCV(rf2, param_grid=grid_values, scoring='r2', cv=cv,verbose=2)
rf_cv.fit(X_train1, y_train)

toc = time.time()

print('time:', round(toc-tic, 2),'s')
```

```python
import matplotlib.pyplot as plt
max_features = rf_cv.cv_results_['param_max_features'].data
R2_scores = rf_cv.cv_results_['mean_test_score']

plt.figure(figsize=(8, 6))
plt.xlabel('max features', fontsize=16)
plt.ylabel('CV R2', fontsize=16)
plt.scatter(max_features, R2_scores, s=30)
plt.plot(max_features, R2_scores, linewidth=3)
plt.grid(True, which='both')
```

```python
print(rf_cv.best_params_)
```

```python
def OSR2(model, X_test, y_test, y_train):
    
    y_pred = model.predict(X_test)
    SSE = np.sum((y_test - y_pred)**2)
    SST = np.sum((y_test - np.mean(y_train))**2)
                 
    return (1 - SSE/SST)
```

```python
# Model evaluation - Random Forest

print('Cross-validated R2:', round(rf_cv.best_score_, 5))
print('OSR2:', round(OSR2(rf_cv, X_test1, y_test, y_train), 5))
```

## Boosting with CV

```python
# Boosting with cross validation

from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from math import sqrt

grid_values = {'n_estimators':[100,500], 
               'learning_rate': [0.1,0.05,0.02],
               'max_depth':[4], 
               'min_samples_leaf':[3], 
               'max_features':[1.0] } 

gbr = GradientBoostingRegressor() 
cv = KFold(n_splits=5,random_state=88,shuffle=True) 
gbr_cv = GridSearchCV(gbr, param_grid=grid_values, 
                      scoring='r2', cv=cv,verbose=1).fit(X_train1, y_train)

y_pred = gbr_cv.predict(X_test1)
```

```python
#VIF function

from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm

def VIF(df, columns):
    values = sm.add_constant(df[columns]).values
    num_columns = len(columns)+1
    vif = [variance_inflation_factor(values, i) for i in range(num_columns)]
    return pd.Series(vif[1:], index=columns)
```

```python
#Define OSR2 function for linear regression
#because the data used in the linear regression model is not train test splitted data. 

def OSR2_linear(model, df_train, df_test, dependent_var):   
    y_test = df_test[dependent_var]
    y_pred = model.predict(df_test)
    SSE = np.sum((y_test - y_pred)**2)
    SST = np.sum((y_test - np.mean(df_train[dependent_var]))**2)    
    return 1 - SSE/SST
```

## Model Comparison through Bootstrapping

```python
#pip install bootstrapped
```

```python
import numpy as np
import bootstrapped.bootstrap as bs
import bootstrapped.stats_functions as bs_stats
```

```python
#Define metric functions to use in bootstrap

def OS_R_squared(predictions, y_test,y_train):
    SSE = np.sum((y_test-predictions)**2)
    SST = np.sum((y_test-np.mean(y_train))**2)
    r2 = 1-SSE/SST
    return r2

def mean_squared_error(predictions, y_test,y_train):
    MSE = np.mean((y_test-predictions)**2)
    return MSE

def mean_absolute_error(predictions, y_test,y_train):
    MAE = np.mean(np.abs(y_test-predictions))
    return MAE
```

```python
#Bootstrap function for model comparison

def bootstrap_validation(test_data, test_label, train_label,
                         model0,model1,model2,model3,model4,
                         metrics_list, sample=500, random_state=66):
    n_sample = sample
    n_metrics = len(metrics_list)
    output_array1=np.zeros([n_sample, n_metrics])
    output_array1[:]=np.nan
    output_array2=np.zeros([n_sample, n_metrics])
    output_array2[:]=np.nan
    output_array3=np.zeros([n_sample, n_metrics])
    output_array3[:]=np.nan
    output_array4=np.zeros([n_sample, n_metrics])
    output_array4[:]=np.nan
    #print(output_array.shape)
    
    for bs_iter in range(n_sample):
        bs_index = np.random.choice(test_data.index, len(test_data.index), replace=True)
        bs_data = test_data.loc[bs_index]
        bs_label = test_label.loc[bs_index]
        bs_data_gbr = bs_data.drop(columns = ['TO%', 'TRB%'])
      
        bs_predicted_bagging = model0.predict(bs_data)
        
        bs_predicted1 = model1.predict(bs_data)
        bs_predicted2 = model2.predict(bs_data)
        bs_predicted3 = model3.predict(bs_data_gbr)
        bs_predicted4 = model4.predict(bs_data)
        
        for metrics_iter in range(n_metrics):
            metrics = metrics_list[metrics_iter]
            output_array1[bs_iter, metrics_iter]=metrics(bs_predicted1,bs_label,train_label)-metrics(bs_predicted_bagging,bs_label,train_label)
            output_array2[bs_iter, metrics_iter]=metrics(bs_predicted2,bs_label,train_label)-metrics(bs_predicted_bagging,bs_label,train_label)
            output_array3[bs_iter, metrics_iter]=metrics(bs_predicted3,bs_label,train_label)-metrics(bs_predicted_bagging,bs_label,train_label)
            output_array4[bs_iter, metrics_iter]=metrics(bs_predicted4,bs_label,train_label)-metrics(bs_predicted_bagging,bs_label,train_label)
            
    output_df1 = pd.DataFrame(output_array1)
    output_df2 = pd.DataFrame(output_array2)
    output_df3 = pd.DataFrame(output_array3)
    output_df4 = pd.DataFrame(output_array4)
    
    return output_df1, output_df2, output_df3, output_df4
```

```python
bs_output1, bs_output2, bs_output3, bs_output4 = bootstrap_validation(X_test1, y_test, y_train,
                                                          bagging_cv, rf_cv, dtr_cv, gbr_cv_feature_selection, linear36,
                                                          metrics_list=[OS_R_squared],
                                                          sample = 5000)
```

```python
#Bootstrapping metrics of the best model

def bootstrap_best(test_data, test_label, train_label, model, 
                   metrics_list, sample=500, random_state=66):
    n_sample = sample
    n_metrics = len(metrics_list)
    output_array=np.zeros([n_sample, n_metrics])
    output_array[:]=np.nan
    print(output_array.shape)
    for bs_iter in range(n_sample):
        bs_index = np.random.choice(test_data.index, len(test_data.index), replace=True)
        bs_data = test_data.loc[bs_index]
        bs_label = test_label.loc[bs_index]
        bs_predicted = model.predict(bs_data)
        for metrics_iter in range(n_metrics):
            metrics = metrics_list[metrics_iter]
            output_array[bs_iter, metrics_iter]=metrics(bs_predicted,bs_label,train_label)
    output_df = pd.DataFrame(output_array)
    return output_df
```
